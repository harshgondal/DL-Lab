```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import torch.optim as optim
```

```python
x = torch.tensor([5.0, 7.0, 12.0, 16.0, 20.0]).view(-1, 1)  # Reshaped for NN input
y = torch.tensor([40.0, 120.0, 180.0, 210.0, 240.0]).view(-1, 1)
print(x)
```

```python
class RegressionModel(nn.Module):
    def __init__(self):
        super(RegressionModel, self).__init__()
        self.w = torch.nn.Parameter(torch.rand([1]))
        self.b = torch.nn.Parameter(torch.rand([1]))
        # print(self.w.item())
    def forward(self, x):
        # print(self.w*x+self.b)
        return self.w*x+self.b
model=RegressionModel()
criterion=nn.MSELoss()
optimizer=optim.SGD(model.parameters(), lr=0.001)

loss_list = []

for epoch in range(100):
    model.train()

    # Forward pass
    y_pred = model(x)
    loss = criterion(y_pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    loss_list.append(loss.item())
```

```python
plt.plot(loss_list)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()

print("Learned weight:", model.w.item())
print("Learned bias:", model.b.item())
```
